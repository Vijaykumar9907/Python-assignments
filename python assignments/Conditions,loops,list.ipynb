```python
#The JEE

student1 = list(map(int, input().split()))
student2 = list(map(int, input().split()))
total_marks1 = sum(student1)
total_marks2 = sum(student2)
if total_marks1 > total_marks2:
      print ("First")
elif total_marks2 > total_marks1:
      print("Second")
else:
      if student1[2] > student2[2]:
           print("First")
      elif student2[2] > student1[2]:
           print("Second")
      else:
          if student1[0] > student2[0]:
              print("First")
          elif student2[0] > student1[0]:
               print("Second")
          else:
              print("Tie")




#Score in 50 overs

N= int(input())
runs=N*50
print(runs)



#Evaluate the Expression

A,B,C,D,E=map(int,input().split())
expression = (A+B)*(C+D+E)
print(expression)



#Runs required in last over

N,n1,n2,n3,n4,n5,n6=list(map(int,input().split()))
runs=N-(n1+n2+n3+n4+n5+n6)
print(runs)




#Pattern Printing - Hollow Square

N=int(input())
for row in range(N):
    for col in range(N):
        if row==0 or row==N-1 or col==0 or col==N-1:
          print("*",end=' ')
        else:
          print(" " ,end=' ') 
    print()




#Sum of Natural Numbers

number=int(input())
Natural_Numbers=(number*(number+1))//2
print(Natural_Numbers)
```
